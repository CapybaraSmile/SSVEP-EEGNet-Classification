# Classification Accuracy with Different Filter Groups

```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
filter_groups = ['filter 1', 'filter 1,2', 'filter 1,3', 'filter 2,3', 'filter 1,2,3']
accuracies = [85, 88, 90, 87, 92]  # 分类准确率 (%)
std_devs = [2, 1.5, 1, 1.8, 1.2]  # 标准差 (%)

# 创建柱状图
x = np.arange(len(filter_groups))  # 横轴位置
plt.bar(x, accuracies, yerr=std_devs, capsize=5, color='skyblue', label='Accuracy (%)')

# 设置横轴标签
plt.xticks(x, filter_groups)
plt.xlabel('Filter Groups')
plt.ylabel('ITR(bits/min)')

# 显示图形
plt.show()
```
