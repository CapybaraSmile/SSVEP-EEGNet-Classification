```python
# 模拟数据
time_windows = [0.3, 0.5, 0.7, 1.0]
accuracy_algo1 = [57, 84, 100, 94]
accuracy_algo2 = [49, 78, 97, 93]
accuracy_algo3 = [42, 65, 81, 77]

# 绘制折线图
plt.figure(figsize=(10, 6))
plt.plot(time_windows, accuracy_algo1, marker='o', label='FB-EEGNet')
plt.plot(time_windows, accuracy_algo2, marker='o', label='EEGNet')
plt.plot(time_windows, accuracy_algo3, marker='o', label='CCA')

# 添加图例、标题和标签
plt.ylim(30, 120)
plt.title('ITR under different window length')
plt.xlabel('Time window length(s)')
plt.ylabel('ITR(bits/min)')
plt.legend()
plt.grid(True)

# 显示图形
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
methods = ['FB-EEGNet', 'EEGNet','CCA']
accuracy = [84, 81, 74]  # 示例准确率数据
itr = [94, 93, 77]  # 示例ITR数据

x = np.arange(len(methods))  # 条形图的位置
width = 0.5  # 柱子的宽度

fig, ax = plt.subplots(2, 1, figsize=(10, 12))  # 创建两行一列的子图

# 准确率条形图
ax[0].bar(x, accuracy, width, color=['purple', 'orange', 'green'])
ax[0].set_ylabel('Accuracy (%)')
ax[0].set_title('Accuracy at 1s Time Window')
ax[0].set_xticks(x)
ax[0].set_xticklabels(methods)
ax[0].set_ylim(40, 100)

# ITR条形图
ax[1].bar(x, itr, width, color=['purple', 'orange', 'green'])
ax[1].set_ylabel('ITR (bits/min)')
ax[1].set_title('ITR at 1s Time Window')
ax[1].set_xticks(x)
ax[1].set_xticklabels(methods)
ax[1].set_ylim(50, 110)

'''for i in range(len(accuracy)):
    ax[0].text(i, accuracy[i] + 0.5, f'{accuracy[i]:.1f}', ha='center')
for i in range(len(itr)):
    ax[1].text(i, itr[i] + 0.5, f'{itr[i]:.1f}', ha='center')'''

plt.tight_layout()
plt.show()
```
