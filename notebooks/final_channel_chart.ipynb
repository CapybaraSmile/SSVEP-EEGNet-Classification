```python
import matplotlib.pyplot as plt
import numpy as np

# 示例数据
filter_groups = ['3-channel', '6-channel', '9-channel', '64-channel']
accuracies = [85, 88, 87, 92]  # 分类准确率 (%)
std_devs = [2, 1.5, 1.8, 1.2]  # 标准差 (%)

# 创建柱状图
x = np.arange(len(filter_groups))  # 横轴位置
plt.bar(x, accuracies, yerr=std_devs, capsize=5, color='#ff7f0e', label='Accuracy (%)')

# 设置横轴标签
plt.xticks(x, filter_groups)
plt.xlabel('Filter Groups')
plt.ylabel('Accuracy(%)')

# 显示图形
plt.show()
```
