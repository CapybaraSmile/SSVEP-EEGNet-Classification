# Accuracy of EEGNet, FB-EEGNet and CCA

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
times = ['0.4', '0.6', '0.8', '1.0']
methods = ['EEGNet', 'FB-EEGNet', 'CCA']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Accuracy
ITR = {
    'EEGNet': [147.71,143.81,142.94,118.01],
    'FB-EEGNet': [187.70,162.69,146.53,129.94],
    'CCA': [18.89,39.98,54.26,61.37]
}

# Standard diveation
std_dev = {
    'EEGNet': [31.54,23.96,21.47,17.56],
    'FB-EEGNet': [26.83,26.79,23.62,19.32],
    'CCA': [10.56,16.90,16.93,17.33]
}
```

```python
n_times = len(times)
n_methods = len(methods)
bar_width = 0.25  # 单根柱子宽度
total_group_width = n_methods * bar_width  # 每组总宽度
```

```python
# 生成基准横坐标（每组左侧起始位置）
x_base = np.arange(n_times)  # [0,1,2,3]

plt.figure(figsize=(10, 6))

for i, method in enumerate(methods):
    # 计算当前方法的横坐标（同一组内连续排列）
    x = x_base + i * bar_width
    
    plt.bar(
        x,
        ITR[method],
        width=bar_width,
        color=colors[i],
        label=method,
        yerr=std_dev[method],
        capsize=5,
        error_kw={'elinewidth':1, 'ecolor':'black'}
    )

# ----------------------
# 4. 调整横坐标标签位置
# ----------------------
# 将标签设置在每组中央
group_centers = x_base + total_group_width / 2
plt.xticks(group_centers, times)

# ----------------------
# 5. 其他装饰（保持原样）
# ----------------------
plt.xlabel('Time(s)', fontsize=12)
plt.ylabel('ITR(bits/min)', fontsize=12)
#plt.title('Accuracy of EEGNet, FB-EEGNet and CCA', fontsize=14)
plt.ylim(0,220)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
